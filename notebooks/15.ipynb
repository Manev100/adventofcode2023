```python
def hash_algorithm(string):
    current_value = 0 
    for character in list(string):
        ascii = ord(character)
        current_value += ascii
        current_value *= 17
        current_value = current_value % 256

    return current_value


assert hash_algorithm("HASH") == 52
```

```python
def solution (data, verbose=False):
    hashes = []
    for step in data:
        hashes.append(hash_algorithm(step))

    return hashes


sol = solution(data)
# sol = solution(test_data, verbose=True)

print(sol)
display(sum(sol))
```

```python
def solution (data, verbose=False):
    boxes = [OrderedDict() for _ in range(256)]

    for step in data:
        if "=" in step:
            label, focal_length = step.split("=")
            label_hash = hash_algorithm(label)
            box = boxes[label_hash]
            if label in box.keys():
                box[label] = int(focal_length)
            else:
                box[label] = int(focal_length)
        elif "-" in step:
            label = step.split("-")[0]
            label_hash = hash_algorithm(label)
            box = boxes[label_hash]
            if label in box.keys():
                box.pop(label)

    focusing_power = []
    for idx, box in enumerate(boxes):
        if len(box) > 0:
            for idx_lens, (label, focal_len) in enumerate(box.items()):
                power = (idx + 1) * (idx_lens + 1) * focal_len
                focusing_power.append(power)

    return focusing_power


sol = solution(data)
# sol = solution(test_data, verbose=True)

print(sol)
display(sum(sol))
```
